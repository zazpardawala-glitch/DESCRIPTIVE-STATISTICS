Variance measures how spread out the data is from the mean.

A smaller variance means the data points (delivery delays) are closer to the average ‚Üí more consistent.

A larger variance means the delays fluctuate more widely ‚Üí less consistent.

Company A‚Äôs variance (6) is lower than Company B‚Äôs variance (15).

This means Company A‚Äôs delivery delays are more stable and predictable, while Company B‚Äôs delays vary more.

For decision-making, a company with lower variance is usually preferred because customers value reliability over occasional fast but inconsistent service.

 Company A is more consistent because its lower variance (6 vs. 15) shows less fluctuation in delivery delays, making its service more reliable.
